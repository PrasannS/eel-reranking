```python
def get_posunique(sentence, noun):
    options = ["VB", "VBP"]
    if noun:
        options = ["NN", "NNP", "NNS"]
    text = nltk.word_tokenize(sentence)
    pos_tagged = nltk.pos_tag(text)
    unwords = set()
    for p in pos_tagged:
        if p[1] in options:
            unwords.add(p[0])
    return len(unwords)

def undset(df, noun):
    res = []
    for index, row in df.iterrows():
        if index%10000==0:
            print(index)
        if "-en" in row['lp']:
            res.append({
                'src':row['mt'],
                'mt':row['mt'],
                'score':get_posunique(row['mt'], noun)
            })
        elif "en-" in row['lp']:
            res.append({
                'src':row['src'],
                'mt':row['src'],
                'score':get_posunique(row['src'], noun)
            })
    return pd.DataFrame(res)

verbdset = undset(traindset, False)
```

```python
verbdset.to_csv("../COMET/data/uniqueverbs.csv")
noundset = undset(traindset, True)
noundset.to_csv("../COMET/data/uniquenouns.csv")
```
