```python
import json
```

```python
def read_file(fname):
    data = []
    with open(fname+'.json', 'r') as myfile:
        data=myfile.read()

    # parse file
    obj = json.loads(data)
    return obj

filenames = ['fillblank', 'formal', 'opposite', 'simplify', 'summary', 'simplifhand']

def print_fname_output (find):
    jdata = read_file('prompt_output/'+filenames[find]+'_prompt')
    #print(jdata['data'][0])
    for data in jdata['data']:
        print("PROMPT -> "+data['prompt'])
        for e in data['examples']:
            print()
            print(e['input'])
            #print(e.keys())
            for d in e['data']:
                print(d['out_toks'])
                print(d['out_sco'])
                
```
